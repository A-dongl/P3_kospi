```python
import pandas as pd

oil = pd.read_csv('DT_oil.csv')

oil.info()
```

```python
oil.fillna(0)
```

```python
oil_1 = oil[oil['거래량'].str.contains('K', na=False)]
oil_1
```

```python
oil_split = oil_1['거래량'].str.split('K')

oil_split2 = oil_1['변동 %'].str.split('%')
oil_ex = oil_split.str.get(0)
oil_ch = oil_split2.str.get(0)

oil_1['거래량'], oil_1['변동 %'] = (oil_ex, oil_ch)
oil_1['거래량'] =oil_1['거래량'].fillna(0)
oil_1['변동 %'] = oil_1['변동 %'].fillna(0)
```

```python
oil_1.drop(['오픈', '고가','저가'], axis=1, inplace=True)
oil_1['거래량']=pd.to_numeric(oil_1['거래량'])
oil_1['변동 %'] = pd.to_numeric(oil_1['변동 %'])
```

```python
oil_1.info()
```

```python
oil_1['거래량'] =oil_1['거래량']*1000
```

```python
oil_2.drop(['오픈', '고가','저가'], axis=1, inplace=True)
oil_2['거래량']=pd.to_numeric(oil_2['거래량'])
oil_2['변동 %'] = pd.to_numeric(oil_2['변동 %'])
```
