```python
# 차분 + expand 모두 사용 or rolling 만
# 추세와 계절성이 있으면 -> rolling, 없으면 -> expanding
# 추세와 계절성을 없애기 위해 -> 차분?

# 결론 -> 특정 값에 진동하지 않음 -> 차분 필요
# 값이 튀는 것들을 보정 -> rolling

dt['cu_c1_wd'] = dt['cu_c1'].rolling(window=7).mean()
dt['cu_c1_wd'].plot(figsize=(12,5)).autoscale(axis='x', tight=True)
```

```python
dt['cu_wd']=dt['cu_cost'].expanding(min_periods=7).mean()
```

```python
from statsmodels.tsa.stattools import adfuller

y = dt['cu_cost']
y_1diff = dt.diff().dropna()['cu_wd']
result = adfuller(y)
print(f'원 데이터 ADF Statistic: {result[0]:.3f}')
print(f'원 데이터 p-value: {result[1]:.3f}')
result = adfuller(y_1diff)
print(f'1차 차분 ADF Statistic: {result[0]:.3f}')
print(f'1차 차분 p-value: {result[1]:.3f}')
```

```python
def interpolated(df):
    temp=[]
    for i in df['wheat_cost'].unique():
        temp.append(df[df['wheat_cost']==i].interpolate())
    new_df=pd.concat(temp,axis=0)
    return new_df
```
