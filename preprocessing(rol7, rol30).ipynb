```python
import pandas as pd
import numpy as np
```

```python
dt1 = pd.read_csv('final2.csv', index_col=0)
dt2 = pd.read_csv('final_cof_bean_pork.csv', index_col=0)
dt = pd.merge(dt1, dt2, left_index=True, right_index=True, how='left')
dt.set_index(keys='date_x', inplace=True)
```

```python
dt3 = pd.read_csv('cof_10month.csv', index_col='date')
dt4 = pd.read_csv('cu_wti_gas_koreastock_10month.csv', index_col='date')
dt3.drop(['Unnamed: 0'], axis=1, inplace=True)
dt3.reset_index()
dt4.reset_index()
dt5 = pd.merge(dt3, dt4, left_index=True, right_index=True, how='left')
```

```python
dt.drop(['kospi_vol', 'kospi200_vol', 'kosdaq_vol','kosdaq100_vol'], axis=1, inplace = True)
dt5.drop(['Unnamed: 0'],axis=1, inplace = True)
```

```python
dt_proc = dt.drop(['kospi_cost', 'kospi200_cost', 'kosdaq_cost','kosdaq100_cost', 'date_y'], axis=1)
dt_proc2 = dt5.drop(['kospi_cost', 'kospi200_cost', 'kosdaq_cost','kosdaq100_cost'], axis=1)
```

```python
dt_proc = dt_proc.interpolate()
dt_proc2 = dt_proc2.interpolate()
dt_proc.sort_values('date_x', inplace=True)
```

```python
dt_proc = pd.concat([dt_proc, dt_proc2], axis=0)
```

```python
dt = dt_proc
```

```python
dt.info()
```

```python
dt['cu_rol7'] = dt['cu_cost'].rolling(window=7).mean()
dt['cu_rol30'] = dt['cu_cost'].rolling(window=30).mean()
dt['cu_rol7_c1'] = dt['cu_rol7'].diff()
dt['cu_rol30_c1'] = dt['cu_rol30'].diff()

dt['oil_rol7'] = dt['oil_cost'].rolling(window=7).mean()
dt['oil_rol30'] = dt['oil_cost'].rolling(window=30).mean()
dt['oil_rol7_c1'] = dt['oil_rol7'].diff()
dt['oil_rol30_c1'] = dt['oil_rol30'].diff()

dt['gas_rol7'] = dt['gas_cost'].rolling(window=7).mean()
dt['gas_rol30'] = dt['gas_cost'].rolling(window=30).mean()
dt['gas_rol7_c1'] = dt['gas_rol7'].diff()
dt['gas_rol30_c1'] = dt['gas_rol30'].diff()

dt['al_rol7'] = dt['al_cost'].rolling(window=7).mean()
dt['al_rol30'] = dt['al_cost'].rolling(window=30).mean()
dt['al_rol7_c1'] = dt['al_rol7'].diff()
dt['al_rol30_c1'] = dt['al_rol30'].diff()

dt['wheat_rol7'] = dt['wheat_cost'].rolling(window=7).mean()
dt['wheat_rol30'] = dt['wheat_cost'].rolling(window=30).mean()
dt['wheat_rol7_c1'] = dt['wheat_rol7'].diff()
dt['wheat_rol30_c1'] = dt['wheat_rol30'].diff()

dt['cof_rol7'] = dt['cof_cost'].rolling(window=7).mean()
dt['cof_rol30'] = dt['cof_cost'].rolling(window=30).mean()
dt['cof_rol7_c1'] = dt['cof_rol7'].diff()
dt['cof_rol30_c1'] = dt['cof_rol30'].diff()

dt['pk_rol7'] = dt['pk_cost'].rolling(window=7).mean()
dt['pk_rol30'] = dt['pk_cost'].rolling(window=30).mean()
dt['pk_rol7_c1'] = dt['pk_rol7'].diff()
dt['pk_rol30_c1'] = dt['pk_rol30'].diff()

dt['bn_rol7'] = dt['bn_cost'].rolling(window=7).mean()
dt['bn_rol30'] = dt['bn_cost'].rolling(window=30).mean()
dt['bn_rol7_c1'] = dt['bn_rol7'].diff()
dt['bn_rol30_c1'] = dt['bn_rol30'].diff()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler_ = MinMaxScaler()
scaled_oil = scaler_.fit_transform(dt[['oil_vol']])
dt['scaled_oil'] = scaled_oil

scaled_oil = scaler_.fit_transform(dt[['cu_vol']])
dt['scaled_cu'] = scaled_oil

scaled_oil = scaler_.fit_transform(dt[['gas_vol']])
dt['scaled_gas'] = scaled_oil

scaled_oil = scaler_.fit_transform(dt[['wheat_vol']])
dt['scaled_wheat'] = scaled_oil

scaled_oil = scaler_.fit_transform(dt[['pk_vol']])
dt['scaled_pk'] = scaled_oil

scaled_oil = scaler_.fit_transform(dt[['al_vol']])
dt['scaled_al'] = scaled_oil

scaled_oil = scaler_.fit_transform(dt[['cof_vol']])
dt['scaled_cof'] = scaled_oil

scaled_oil = scaler_.fit_transform(dt[['bn_vol']])
dt['scaled_bn'] = scaled_oil
```
